# Gráficas interactivas

Para hacer gráficas interactivas necesitamos la biblioteca `interact`. Además, debido a algunos conflictos con la funcionalidad nativa, hemos creado funciones que calculan los polinomios de Hermite, Laguerre y Jacobi utilizando la fórmula de Rodrigues.

```python
from sage.repl.ipython_kernel.interact import interact

def Rodrigues_H(n):
    ''' Calcula el polinomio de Hermite de grado n utilizando
        la formula de Rodrigues
    '''
    Bn = (-1)**n
    rho_inv = e**(x**2)
    der = derivative(e**(-x**2),x,n)
    return (Bn*rho_inv*der).full_simplify()

def Rodrigues_L(n, alpha):
    ''' Calcula el polinomio de Laguerre de grado n y parametro 
        'alpha' utilizando la formula de Rodrigues
    '''
    Bn = 1/factorial(n)
    rho_inv = pow(x,-alpha)*e**x
    der = derivative(x**(n+alpha)*e**(-x),x,n)
    return (Bn*rho_inv*der).full_simplify()

def Rodrigues_J(n, alpha, beta):
    ''' Calcula el polinomio de Jacobi de grado n y parametros 
        'alpha' y 'beta' utilizando la formula de Rodrigues
    '''
    Bn = (-1)**n/(2**n*factorial(n))
    rho_inv = pow(1-x, -alpha)*pow(1+x, -beta)
    der = derivative(pow(1-x, n+alpha)*pow(1+x, n+beta),x,n)
    return Bn*rho_inv*der
```

Con los siguientes comandos creamos las gráficas interactivas

## Hermite

```python
@interact
def Hermite_plot(n=(0,10), a=(-10,0,0.1), b=(0,10,0.1)):
    pol = Rodrigues_H(n)
    plot(pol, a, b, color='red').show()
```

## Laguerre

```python
@interact
def Laguerre_plot(n=(0,10), alpha=(0,10,0.1), a=(0,20,0.1), b=(1,50,0.1)):
    pol = Rodrigues_L(n, alpha)
    plot(pol, x, a, b, color='red').show()
```

## Jacobi

```python
@interact
def plot_Jacobi(n=(0,10), alpha=(0,10,0.1), beta=(0,10,0.1), a=(-1,0,0.1), b=(0,1,0.1)):
    pol = Rodrigues_J(n, alpha, beta)
    plot(pol, x, a, b, color='red').show()
```

# Laguerre

Continuamos con los polinomios de Laguerre. Esta familia de polinomios cuenta con un parámetro $\alpha > -1$. Para poder observar el impacto de este parámetro en las gráficas, fijaremos un grado determinado y representaremos gráficas para distintos valores de $\alpha$.

En la siguiente gráfica vemos los polinomios de Laguerre de grado $2$ con distintos valores no negativos de $\alpha$.

```python
P = plot([]) # Gráfica inicialmente vacía
colors = ['red', 'green', 'blue', 'black', 'brown', 'orange'] # Colores de las curvas
varsalpha = [0, 1, 5, 15, 20, 25]
for j, alpha, color in zip(range(6), varsalpha , colors):
    pol = Rodrigues_L(2,alpha)
    P += plot(pol, 0, 60, color=color, legend_label='$L_{' + str(2) + '}^{(' + '%.1f' % alpha + ')}(x)$')
P.set_legend_options(font_size=legend_size, bbox_to_anchor=(1, 1, 0, 0))
P.save("../memoria/img/C2/Laguerre1.png")
P.show()
```
