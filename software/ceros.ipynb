El objetivo es crear una imagen con las gráficas de dos o más polinomios de una SPO (en este caso concreto de los polinomios de Tchebichef) señalizando sus ceros.

Para ello, `sage` cuenta con funciones como:
* `plot`: Representa gráficamente la función de `x` que se desee, especificando también los extremos inferior y superior del intervalo que se desea representar. También es posible especificar una etiqueta para la leyenda del gráfico y el color que queremos que se visualice la curva.

```python
ext_inf, ext_sup = -1, 1 # Extremos del intervalo 
plot(chebyshev_T(6,x), ext_inf, ext_sup, color = 'blue')
```

* `point`: Muestra un punto 2D a partir de sus coordenadas, pudiendo especificar el tamaño y el color de la representación.

```python
point((1,0), pointsize = 50, color = 'red')
```

Por otro lado, la 'suma' de gráficas superpone las gráficas en una sola:

```python
plot(chebyshev_T(6,x), ext_inf, ext_sup, color = 'blue') + \
point((0,0), pointsize = 50, color = 'red')
```

Con estos ingredientes, podemos crear un código que muestre algunas gráficas de polinomios y sus raíces:

```python
P = plot([]) # Gráfica inicialmente vacía
colors = ['red', 'green', 'blue', 'black'] # Colores de las curvas
for j, color in zip(range(2,6), colors): # Representamos T2, T3, T4 y T5
    pol = chebyshev_T(j,x) # j-ésimo polinomio de Tchebichef
    
    # Representación de la curva, añadimos una etiqueta para la leyenda
    P += plot(pol, ext_inf, ext_sup, color=color, legend_label='$T_{' + str(j) + '}(x)$')
    
    # Calculamos los ceros del polinomio
    zeros = [root[0] for root in pol.roots()]
    # Representamos estos puntos con su color correspondiente
    for p in zeros:
        P += point((p,0), pointsize=50, color=color)    
# Mostramos la gráfica situando la leyenda en una esquina    
P.show(legend_loc="upper right" )
```
